```python
#Combining the added constraint (61 total fast-tracks) with the goal of beating the baseline (41 correct predictions), we can formulate the new objective as: find a classifier that has a recall@k (with k=61) greater than 41.
```

```python
from sklearn.model_selection import train_test_split

feature = df.drop(columns='had_heart_attack')
target = df[['had_heart_attack']]

X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size = 0.2, random_state=42)
```

```python
# calculate VIF scores
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif 
from statsmodels.tools.tools import add_constant

X = add_constant(X_train)
X
```

```python
vif_df = pd.DataFrame([vif(X.values, i)  for i in range(X.shape[1])], index=X.columns).reset_index()
vif_df.columns = ['feature','vif_score']
vif_df = vif_df.loc[vif_df.feature!='const']
vif_df
```

```python
# heatmap correlation
price_train = pd.concat([X_train, y_train], axis=1)
corr = price_train.corr()

plt.figure(figsize=(10,7))
sns.heatmap(corr, annot=True, fmt='.2f')
plt.show()
```

```python
# drop redundant features
X_train = X_train.drop(columns=['oldpeak', 'exng'])
X_test = X_test.drop(columns=['oldpeak','exng'])
```

```python
# heatmap correlation
price_train = pd.concat([X_train, y_train], axis=1)
corr = price_train.corr()

plt.figure(figsize=(10,7))
sns.heatmap(corr, annot=True, fmt='.2f')
plt.show()
```

```python
# Checking the VIF score again

X = add_constant(X_train)

vif_df = pd.DataFrame([vif(X.values, i) for i in range(X.shape[1])],  index=X.columns).reset_index()
vif_df.columns = ['feature','vif_score']
vif_df = vif_df.loc[vif_df.feature!='const']
vif_df
```

```python
# train the model
X_train = X_train.to_numpy()
y_train = y_train.to_numpy()
y_train = y_train.reshape(len(y_train),)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler().fit(X_train)
```

```python
X_scaled = scaler.transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(max_iter = 200, random_state=0)
```

```python
model = clf.fit(X_scaled, y_train)
```

```python
model.score(X_test, y_test)
```

```python
X = X.drop(['const'], axis = 1)
X
```

```python
X['pred'] = model.predict(X)
```

```python
with_preds = X.sort_values('pred').tail(61)
```

```python
top_k_preds = pd.concat([with_preds, y_train],  axis=1)
```

```python
top_k_preds
```

```python
recall_at_k = len(
  top_k_preds
  .query('had_heart_attack == 1')
  .query('pred == 1'))
```

```python
print(f"Recall@61 = {recall_at_k}")
```
